```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Indian automoble buying behavour study 1.0.csv")
df
```

###### Analyze age groups interested in EVs.

```python
df["Age"].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=40, kde=True, color='skyblue')
plt.title('Demographic Segmentation - Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

###### Explore professions of potential EV buyers

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Profession', palette='viridis')
plt.title('Psychographic Segmentation - Profession')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

###### Analyze the behavior of individuals with respect to personal loans and house loans

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Personal loan', hue='House Loan')
plt.title('Behavioral Segmentation - Personal Loan vs. House Loan')
plt.xlabel('Personal Loan')
plt.ylabel('Count')
plt.show()
```

###### Age vs. Salary

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Profession', hue='Personal loan', data=df)
plt.title('Profession vs. Personal Loan')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Education', y='No of Dependents', hue='House Loan', data=df)
plt.title('Education, No of Dependents, and House Loan')
plt.xlabel('Education')
plt.ylabel('No of Dependents')
plt.show()
```

###### the distribution of car makes among different professions

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Profession', hue='Make', data=df)
plt.title('Profession vs. Make')
plt.xlabel('Profession')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Make')
plt.show()
```

###### the distribution of car prices for each make.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Make', y='Price', data=df)
plt.title('Make vs. Price')
plt.xlabel('Make')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Profession', y='Price', data=df)
plt.title('Profession vs. Price')
plt.xlabel('Profession')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Education', y='Salary', data=df)
plt.title('Education vs. Salary')
plt.xlabel('Education')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Education', hue='Make', data=df)
plt.title('Education vs. Make')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Make', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

```python
numerical_cols = ['Age', 'Salary', 'Total Salary']

# Create a pairplot
sns.set(style="ticks")
sns.pairplot(df, vars=numerical_cols, hue='Make', diag_kind='kde', markers='o', palette='husl')
plt.suptitle('Multivariate Analysis of Age, Salary, and Total Salary by Make', y=1.02)
plt.show()
```
